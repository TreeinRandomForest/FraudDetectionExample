### Split data into train and test

Most machine learning models overfit i.e. memorize the dataset given enough data. To hedge against this, we will split the dataset into two disjoint pieces - the **train** and the **test** set. The train set is used to train or build the model and the test set is used to evaluate how well the model does.

If all the samples in the dataset are independent from each other, then one can randomly choose, say 75% of the dataset and define it to be the train set and define the rest to be the test set.

When time is a factor (like the **Time** column), then there's a possibility that the samples are **not independent**. In this case, while we should carry out transformation on variables to de-correlate them, we should also split the train and test sets by time. More precisely, we pick a Time value and define everything before to be train and everything after to be test.

So while the ratio between Class 0 in train vs test ~ 3, as we would expect given the 75%-25% split, the ratio between Class 1 ~ 4.25.

This indicates there are relatively more Class 1 samples in train than in test.

**Aside**: To first approximation, we can treat the number of samples in each Class to be sampled from a Poisson distribution which models counts. A Poisson distribution has one parameter (mean count) with standard deviation = $\sqrt{\text{mean}}$.

In this case, the number of Class 1 samples in train can be treated as a Poisson variable with mean, $\mu = 398$ and standard deviation $\sigma=\sqrt{\mu}=\sqrt{398}$. Similarly, the number of Class 1 samples in test can be treated as a Poisson with mean, $\mu = 94$ and standard deviation, $\sigma = \sqrt{\mu} = \sqrt{94}$.

We can use this to measure the 1-deviation error in the ratio.

```python
#increase both variables by 1 standard deviation
s1 = 1
s2 = 1

ratio = (398 + s1 * np.sqrt(398)) / (94 + s2 * np.sqrt(94))
print(f'ratio = {ratio}')

#decrease train by 1 std dev and increase test by 1 std dev
s1 = -1
s2 = 1

ratio = (398 + s1 * np.sqrt(398)) / (94 + s2 * np.sqrt(94))
print(f'ratio = {ratio}')

#increase train by 1 std dev and decrease test by 1 std dev
s1 = 1
s2 = -1

ratio = (398 + s1 * np.sqrt(398)) / (94 + s2 * np.sqrt(94))
print(f'ratio = {ratio}')

#decrease both by 1 std dev
s1 = -1
s2 = -1

ratio = (398 + s1 * np.sqrt(398)) / (94 + s2 * np.sqrt(94))
print(f'ratio = {ratio}')

#decrease train by 2 std dev and increase test by 2 std dev
s1 = -2
s2 = 2

ratio = (398 + s1 * np.sqrt(398)) / (94 + s2 * np.sqrt(94))
print(f'ratio = {ratio}')
```

So if we were to vary each count by 2 standard deviations (numerator fluctuates down and denominator fluctuates up), then we would get close to the factor of 3. This is reasonably likely statistically (high p-value) and so to first approximation, we won't add any corrective factors for the change in distribution of the Class variable.

#### Evaluating the model

This is a binary classification problem i.e. we are predicting whether a sample is in Class 0 or Class 1. Most models predict a probability of being in Class 1.

**Technical point**: 

Most models predict a score between 0 and 1 with higher scores being more likely to be in Class 1. This score is technically not a probability unless calibrated properly. More concretely, if two different models, say a Random Forest and a k-Nearest Neighbors algorithm, produce two scores for a sample, they cannot be compared directly without calibration.

**End of technical point**

At the simplest level, one can pick a threshold between 0 and 1, label every sample with score > threshold as 1 and every sample with score <= threshold as 0 and compare the actual Class labels with the predictions. A common way would be accuracy i.e. % of predictions gotten correct.

There are two problems with this approach:

1. Dependence on an arbitrary threshold.

2. Accuracy is a misleading metric when they classes are so imbalanced. In our case, a model that predicts Class 0 for every sample (every sample is normal) would have an accuracy of 99.86% on the test set since there are only 94 Class 1 samples and 71108 Class 0 samples. Clearly, this is not what we want.

**In order to better probe the model, we use metrics like area under the ROC curve (AUC), precision and recall.**

#### Precision-recall curve

Both values closer to 1 (top-right corner) are better. There's a tradeoff between higher recall (right on the x-axis) and higher precision (up on the y-axis).

```python
#precision-recall curve
precision_train, recall_train, thresholds_train = precision_recall_curve(target_train, pred_train[:,1])
precision_test, recall_test, thresholds_test = precision_recall_curve(target_test, pred_test[:,1])

plt.figure(figsize=(10,10))
plt.plot(recall_train, precision_train, label='train')
plt.plot(recall_test, precision_test, label='test')
plt.xlabel('recall')
plt.ylabel('precision')
plt.legend()
```

#### Summarize above curve by a single score - area under the curve (ideally = 1, higher is better)

```python
auc_pr_train = average_precision_score(target_train, pred_train[:,1], average=None)
auc_pr_test = average_precision_score(target_test, pred_test[:,1], average=None)

print(f'Train score = {auc_pr_train}')
print(f'Test score = {auc_pr_test}')
```

#### Account for class imbalances while training

We can also make our classifier take the class imbalance into account. Each sample from Class 1 is weighted by a factor much larger than 1 to make it count as much as a sample from Class 0.

### Helper function to train and evaluate models

```python
def build_model(model, features_train, target_train, features_test, target_test):
    '''Helper function to train model
    '''
    model.fit(features_train, target_train)

    pred_train = model.predict_proba(features_train)
    pred_test = model.predict_proba(features_test)

    #pr curve
    precision_train, recall_train, thresholds_train = precision_recall_curve(target_train, pred_train[:,1])
    precision_test, recall_test, thresholds_test = precision_recall_curve(target_test, pred_test[:,1])

    #roc curve
    fpr_train, tpr_train, _ = roc_curve(target_train, pred_train[:,1])
    fpr_test, tpr_test, _ = roc_curve(target_test, pred_test[:,1])
    
    #area under pr curve
    auc_pr_train = average_precision_score(target_train, pred_train[:,1], average=None)
    auc_pr_test = average_precision_score(target_test, pred_test[:,1], average=None)    
    
    #auc roc
    auc_train = roc_auc_score(target_train, pred_train[:,1])
    auc_test = roc_auc_score(target_test, pred_test[:,1])
    
    plt.figure(figsize=(10,16))
    
    plt.subplot(211)
    plt.plot(recall_train, precision_train, label='train')
    plt.plot(recall_test, precision_test, label='test')
    plt.xlabel('recall')
    plt.ylabel('precision')
    plt.legend()
    plt.title(f'Train Area = {auc_pr_train: .3f} - Test Area = {auc_pr_test: .3f}')
 
    plt.subplot(212)
    plt.plot(fpr_train, tpr_train, label='train')
    plt.plot(fpr_test, tpr_test, label='test')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.title(f'Train Area = {auc_train: .3f} - Test Area = {auc_test: .3f}')    
    
    return model
```

```python
class Ensembler:
    def __init__(self, estimator_list, df_train, df_val, df_test, target_colname):
        self.estimator_list = estimator_list
        
        self.df_train = df_train
        self.df_val = df_val
        self.df_test = df_test
        
        self.target_colname = target_colname
        
    def construct_new_data(self, df, keep_orig_features=False):
        features = df.drop(self.target_colname, axis=1)
        
        preds = []
        for est in self.estimator_list:
            preds.append(est.predict_proba(features)[:,1]) #binary classification for now
            
        df_ensemble = pd.DataFrame(np.vstack(preds).transpose(), index=df.index)
        
        if keep_orig_features:
            df_ensemble = pd.concat([df, df_ensemble], axis=1)
        else:
            df_ensemble[self.target_colname] = df[self.target_colname]
            
        return df_ensemble
    
    def prepare_datasets(self, keep_orig_features=False):
        
        df_train_ensemble = self.construct_new_data(self.df_train, keep_orig_features=keep_orig_features)
        
        df_val_ensemble = self.construct_new_data(self.df_val, keep_orig_features=keep_orig_features)
        
        df_test_ensemble = self.construct_new_data(self.df_test, keep_orig_features=keep_orig_features)
        
        return (df_train_ensemble, df_val_ensemble, df_test_ensemble)
```

## Helper Functions

```python
def plot_single_variable_histogram(df, *, column_name='V1', binning_method=10):
    _, bins = np.histogram(df[column_name].tolist(), bins=binning_method)
        
    plt.figure(figsize=(14,5))
    
    plt.subplot('121')
    plt.hist(df[df['Class']==0][column_name].tolist(), bins=bins, label = '0')
    plt.legend()
    plt.title(f'Normal Transactions - Column = {column_name}')
    plt.yscale('log')
    
    plt.subplot('122')
    plt.hist(df[df['Class']==1][column_name].tolist(), bins=bins, label = '1')
    plt.legend()
    plt.title(f'Fraudulent Transactions - Column = {column_name}')
    plt.yscale('log')

def preprocess_pca(df, pca_cols, *, pca_n_components=2):
    df_pca = df[pca_cols]

    scaler = StandardScaler(with_mean=True, with_std=True)
    df_pca = scaler.fit_transform(df_pca)

    pca = PCA(n_components=2)
    df_pca = pca.fit_transform(df_pca)    

    df_pca = pd.DataFrame(df_pca)
    df_pca['Time'] = df['Time']
    df_pca['Class'] = df['Class']
    
    return df_pca, scaler, pca
    
def create_validation_sets(df, train_size=0.75):
    #sort dataframe by Time
    df.sort_values(by='Time', inplace=True)

    n_samples = df.shape[0]

    if train_size < 0 or train_size > 1:
        raise ValueError('Please ensure train_size is between 0 and 1')

    n_train = int(train_size * n_samples)

    df_train = df.iloc[0:n_train] #first n_train rows are train
    df_test = df.iloc[n_train:] #rest is test

    return df_train, df_test
    
def build_model(model, features_train, target_train, features_test, target_test):
    '''Helper function to train model
    '''
    if 'Class' in features_train or 'Class' in features_test:
        raise ValueError("Column Class shouldn't be in features")
    
    model.fit(features_train, target_train)

    pred_train = model.predict_proba(features_train)
    pred_test = model.predict_proba(features_test)

    #pr curve
    precision_train, recall_train, thresholds_train = precision_recall_curve(target_train, pred_train[:,1])
    precision_test, recall_test, thresholds_test = precision_recall_curve(target_test, pred_test[:,1])

    #roc curve
    fpr_train, tpr_train, _ = roc_curve(target_train, pred_train[:,1])
    fpr_test, tpr_test, _ = roc_curve(target_test, pred_test[:,1])
    
    #area under pr curve
    auc_pr_train = average_precision_score(target_train, pred_train[:,1], average=None)
    auc_pr_test = average_precision_score(target_test, pred_test[:,1], average=None)    
    
    #auc roc
    auc_train = roc_auc_score(target_train, pred_train[:,1])
    auc_test = roc_auc_score(target_test, pred_test[:,1])
    
    plt.figure(figsize=(10,16))
    
    plt.subplot(211)
    plt.plot(recall_train, precision_train, label='train')
    plt.plot(recall_test, precision_test, label='test')
    plt.xlabel('recall')
    plt.ylabel('precision')
    plt.legend()
    plt.title(f'Train Area = {auc_pr_train: .3f} - Test Area = {auc_pr_test: .3f}')
 
    plt.subplot(212)
    plt.plot(fpr_train, tpr_train, label='train')
    plt.plot(fpr_test, tpr_test, label='test')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.title(f'Train Area = {auc_train: .3f} - Test Area = {auc_test: .3f}')    
    
    return model

def plot_confusion_matrix(train_labels, train_pred):
    fig = plt.figure(figsize=(6,6))
    ax = plt.subplot()

    labels = list(train_labels.value_counts().index)

    confusion = confusion_matrix(train_labels, train_pred, labels=labels)
    ax.matshow(np.log(confusion + 1.001))

    ax.set_xticks(range(len(labels)))
    ax.set_yticks(range(len(labels)))

    ax.set_xticklabels(labels, rotation=90);
    ax.set_yticklabels(labels);

    for i in range(len(labels)):
        for j in range(len(labels)):        
            ax.text(j, i, confusion[i,j], va='center', ha='center')

    plt.xlabel('predicted')    
    plt.ylabel('true')
    
    return fig

def plot_score_dist(model, features, target):
    fig = plt.figure(figsize=(10,10))
    
    n,b,p = plt.hist((model.predict_proba(features)[:,1]))
    plt.clf()
    _ = plt.hist((model.predict_proba(features)[:,1])[np.where(target==0)[0]], stacked=True, 
                                                                               histtype='step',
                                                                               bins=b, 
                                                                               color='g',
                                                                               label='Normal')

    _ = plt.hist((model.predict_proba(features)[:,1])[np.where(target==1)[0]], stacked=True, 
                                                                               histtype='step',
                                                                               bins=b, 
                                                                               color='r',
                                                                               label='Normal')

    plt.yscale('log')
    plt.legend()    
    
    return fig
```
